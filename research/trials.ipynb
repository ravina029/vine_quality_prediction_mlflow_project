```python
#ensure_annotation
from ensure import ensure_annotations
```

```python
def mul(x:int,y:int)->int:
    return x*y

mul(2,3)
```

```python
mul(4,"5") 
##analyse its output you will see the input are not int but still we area getting ans. 
# to avoid this kind of problems we will use ensure_annotations function
```

```python
@ensure_annotations

def mul(x:int,y:int)->int:
    return x*y
```

```python
mul(2,"3")
```
